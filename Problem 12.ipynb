```python
import pandas as pd
import re

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# TRAINING DATASET
data = {
    "complaint": [
        "Incorrect bill charged this month",
        "Billing issue in my invoice",
        "Refund not processed",
        "Payment deducted twice",
        
        "Internet service is very slow",
        "Customer support not responding",
        "Poor service quality",
        "Delayed response from support team",
        
        "App is crashing frequently",
        "Unable to login to account",
        "Website not working properly",
        "Technical issue with application"
    ],
    
    "category": [
        "billing","billing","billing","billing",
        "service","service","service","service",
        "technical","technical","technical","technical"
    ]
}

df = pd.DataFrame(data)

# TEXT PREPROCESSING
def preprocess(text):
    
    text = text.lower()
    
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    return text

df["cleaned_complaint"] = df["complaint"].apply(preprocess)

# TF-IDF VECTORIZATION
vectorizer = TfidfVectorizer(
    stop_words='english'
)

X = vectorizer.fit_transform(
    df["cleaned_complaint"]
)

y = df["category"]

# TRAIN MODEL
model = LogisticRegression()

model.fit(X, y)

# DEPARTMENT MAPPING
department_map = {
    "billing": "Finance Team",
    "service": "Customer Service Team",
    "technical": "Technical Support Team"
}

# USER INPUT / PREDICTION
user_complaint = input("Enter Customer Complaint: ")

cleaned_input = preprocess(user_complaint)

input_vec = vectorizer.transform([cleaned_input])

prediction = model.predict(input_vec)[0]

department = department_map[prediction]

print("\nPredicted Category:", prediction)
print("Assigned Department:", department)
```
